```python
def merge_and_sum_dictionaries(dict1, dict2):
    """
    Merges two dictionaries. If a key exists in both, their values are summed.
    If a key exists in only one, its value is kept.
    The final dictionary is sorted by key.
    """
    merged_dict = dict1.copy() # Start with a copy of the first dictionary

    # Iterate through the second dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            # Key exists in both: add the values
            merged_dict[key] += value
        else:
            # Key exists only in dict2: add the key-value pair
            merged_dict[key] = value

    # Sort the final dictionary by key (ascending alphabetical/dictionary order)
    # The sorted() function returns a list of (key, value) tuples,
    # so we convert it back to an OrderedDict or a regular dict (in Python 3.7+ it preserves order)
    sorted_keys = sorted(merged_dict.keys())
    
    # Create the final sorted dictionary
    final_dict = {key: merged_dict[key] for key in sorted_keys}
    
    return final_dict
```
