```python
input_string=input()
    # Initialize an empty dictionary
result = {}
    
    # Split the input string into key-value pairs
pairs = input_string.split()
    
for pair in pairs:
    key, value = pair.split(':')
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
    
print(result)
```

```python
a=[i for i in map(int,input().split())]
result={}
result["negatives"]=0
result["positives"]=0
result["zeroes"]=0

for i in a:
	if i<0:
		result["negatives"]+=1
	elif i>0:
		result["positives"]+=1
	else: 
		result["zeroes"]+=1

print(result)
```
