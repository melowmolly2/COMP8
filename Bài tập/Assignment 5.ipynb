```python
def Max_Of_Two(a,b)->int:
	if a>b:
		return a
	return b

a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

W5A3

```python
def prime(n):
	if n<2: return False

	for i in range (2,n//2+1):
		if n%i==0:
			return False
		
	return True

n=int(input())
print(prime(n))
```

W5A4

```python
def perfect(n):
	sum=0
	for i in range(1,n//2+1): 
		if n%i==0:
			sum+=i
	
	return sum==n

n=int(input())
print(perfect(n))
```

W5A5

```python
def find(a,k):
	for i in range(0,len(a)):
		if a[i]==k:
			return i+1
	return -1

a=[i for i in map(int,input().split())]
k=int(input())

print(find(a,k))

```

W5A6

```python
def factorial(n):
	if n<=1:
		return 1
	return n*factorial(n-1)

n=int(input())
print(factorial(n))
```

```python
a,b=input().split()
def main(a,b):
	dict1=dict()
	dict2=dict()
	if len(a)!=len(b):
		return False
	for i in range(len(a)):
		if (a[i] in dict1 and b[i]!=dict1[a[i]]) or (b[i] in dict2 and a[i]!=dict2[b[i]]):
			return False
		else:
			dict1[a[i]]=b[i]
			dict2[b[i]]=a[i]
	return True	
print(main(a,b))
```
